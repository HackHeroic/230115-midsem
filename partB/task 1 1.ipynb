## Step-by-Step Method Description (Lee et al., NeurIPS 2010)

Step 1: Surface-geometry cue extraction from image
- The method starts from image evidence and predicts two cues: orientation map and geometric context.
- Reference: Section 3, Figure 2(c,d).
- Purpose: Provide local evidence used later for scoring candidate scene configurations.

Step 2: Generate room hypotheses
- Candidate room layouts are generated as parametric box models using line segments and vanishing-point-aligned geometry.
- Reference: Section 4.1, Figure 2(e).
- Purpose: Build plausible room structures to search over instead of committing to a single early estimate.

Step 3: Generate object hypotheses (cuboids)
- Candidate clutter objects are approximated as cuboids by finding convex-edge-consistent region pairs in orientation maps.
- Reference: Section 4.2, Figure 3(ii).
- Purpose: Represent occupied 3D volume of clutter so object-room interaction can be reasoned physically.

Step 4: Enforce volumetric compatibility constraints
- Candidate combinations are filtered using finite-volume, spatial-exclusion, and containment constraints.
- Reference: Section 2 and Section 4.3, Figure 3(i).
- Purpose: Remove physically invalid scene configurations before final scoring.

Step 5: Structured scene inference with beam search
- The scene variable y=(yr,yo) is inferred by maximizing a learned score while controlling combinatorial growth via beam search.
- Reference: Section 5.1 (inference), equation for y* = arg max_y f(x,y).
- Purpose: Efficiently find strong room+object configuration in large hypothesis space O(n*2^m).

Step 6: Score with appearance agreement plus volumetric penalty
- The score function uses feature agreement and volumetric penalty: f(x,y)=w^T psi(x,y)+w_phi^T phi(y).
- Reference: Section 5.2.
- Purpose: Reward image-consistent layouts and penalize constraint-violating geometry.

Step 7: Learn parameters and evaluate with pixel error
- Weights are learned with structured prediction objective; performance is evaluated with pixel disagreement on room surfaces.
- Reference: Section 5.2 (learning), Section 6 (evaluation), Table 1.
- Purpose: Quantify whether volumetric reasoning improves room layout estimation.

Final summary sentence:
This paper solves single-image indoor room layout estimation under clutter, and claims better performance than prior 2D clutter reasoning by introducing explicit 3D volumetric object-room constraints with structured inference.
