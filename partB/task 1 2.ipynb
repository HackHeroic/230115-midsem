## Assumption 1
Assumption:
Objects can be approximated as cuboids for volumetric reasoning.

Why the method needs it:
The algorithm needs a low-parameter 3D model for clutter to compute room-object containment and object-object overlap. Cuboids provide a tractable geometric proxy for occupied volume and keep inference manageable.

Violation scenario:
A scene dominated by non-cuboidal objects (bean bags, curved furniture, hanging lamps) can produce poor cuboid fits and misleading volumetric constraints.

Paper reference:
Section 4.2, Figure 3(ii), and discussion around object hypotheses as cuboids.

## Assumption 2
Assumption:
Objects rest on the floor plane to resolve scale ambiguity in single-image 3D reasoning.

Why the method needs it:
From one image, absolute scale is ambiguous; assuming floor contact allows consistent compatibility checks between room and object hypotheses in 3D.

Violation scenario:
Wall-mounted shelves, ceiling lamps, suspended objects, or objects on elevated platforms violate this assumption and break compatibility logic.

Paper reference:
Section 4.3 (explicitly states assumption that objects rest on floor).

## Assumption 3
Assumption:
Vanishing-point and orientation-map estimates are reliable enough to seed valid room/object hypotheses.

Why the method needs it:
Hypothesis generation depends directly on orientation cues and Manhattan-structure geometry. If those cues are unstable, downstream hypothesis pool quality degrades before learning can recover it.

Violation scenario:
Images with weak line structure, strong perspective distortions, heavy occlusion, or non-Manhattan architecture can cause VP errors and incorrect layout proposals.

Paper reference:
Section 2, Section 3, Section 4.1; failure discussion in Figure 6.

## Assumption 4
Assumption:
A subset of object hypotheses is sufficient for meaningful volumetric correction of room layout.

Why the method needs it:
The approach does not require perfect object reconstruction; it relies on enough valid cuboids to push/pull room boundaries via containment constraints.

Violation scenario:
If important clutter objects are missed or badly localized, volumetric cues become sparse/noisy and layout gains over baseline may disappear.

Paper reference:
Section 6 qualitative discussion and Figure 4/Figure 6.
