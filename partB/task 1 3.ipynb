Main baseline / prior method:
The primary baseline discussed in the paper is 2D clutter reasoning for room layout, specifically the approach of Hedau et al. (ICCV 2009), which reasons in image plane without explicit 3D object-room interaction.

Limitation of this baseline:
The paper argues that image-plane reasoning can fit room boundaries onto object surfaces in cluttered scenes, because it does not model physically occupied object volume in 3D. As a result, multiple ambiguous room hypotheses may remain plausible from appearance alone.

How the proposed method overcomes this:
Lee et al. add explicit volumetric constraints (containment and spatial exclusion) with cuboid object models and use structured inference over joint room-object configurations, so physically invalid layouts are penalized or rejected.

Condition where the proposed method may not outperform baseline:
If vanishing points and object cuboids are both inaccurate (for example, severe clutter plus weak line structure), volumetric constraints can be noisy or wrong; in that case the added 3D reasoning may not improve and can sometimes hurt relative to a stable 2D baseline.
